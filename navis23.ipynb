```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import CategoricalNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
df = pd.read_csv("playing_tennis.csv")
```

```python
le_outlook=LabelEncoder()
le_temp=LabelEncoder()
le_humidity=LabelEncoder()
le_wind=LabelEncoder()
le_play=LabelEncoder()

df["Outlook"]    = le_outlook.fit_transform(df["Outlook"])
df["Temperature"]= le_temp.fit_transform(df["Temperature"])
df["Humidity"]   = le_humidity.fit_transform(df["Humidity"])
df["Wind"]       = le_wind.fit_transform(df["Wind"])
df["Play"]       = le_play.fit_transform(df["Play"])

```

```python
X = df[["Outlook", "Temperature", "Humidity", "Wind"]]
y = df["Play"]
```

```python
model=CategoricalNB()
model.fit(X,y)
```

```python
y_pred = model.predict(X)
```

```python
print("Accuracy :", accuracy_score(y,y_pred))
print("\nConfusion Matrix :")
print(confusion_matrix(y , y_pred))

print("\nClassification Report :")
print(classification_report(y,y_pred, target_names=le_play.classes_))

print("-" * 50)
```

```python
print("Enter values for prediction: ")

outlook = input("Outlook (Sunny/Overcast/Rain): ")
temperature= input("Temperature (Humidity/Mild/Cool): ")
humidity = input("Humidity (High /Normal): ")
wind = input ("Wind (Weak/strong): ")

user_input =[[
    le_outlook.transform([outlook])[0],
    le_temp.transform([temperature])[0],
    le_humidity.transform([humidity])[0],
    le_wind.transform([wind])[0]
]]

prediction = model.predict(user_input)
probabilities = model.predict_proba(user_input)

final_prediction = le_play.inverse_transform(prediction)

print("\nPrediction : ", final_prediction[0])
print("Probability for Yes :", probabilities[0][le_play.transform(["Yes"])[0]])
print("Probability for No :",probabilities[0][le_play.transform(["No"])[0]])
```
